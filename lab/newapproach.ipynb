# Trying a new appraoch to generate synthetic networks

```python
import networkx as nx

# creating a barabasi albert graph
graph = nx.barabasi_albert_graph(n=10, m=3, seed=42)
nx.draw_networkx(G=graph)
```

## adding synthetic names to the graph using faker
- NB: add this module by running the following command:
`pip install faker`
- or for MacOS people like myself:
`python3 -m pip install faker`

```python
import faker

fake = faker.Faker()
names = [fake.name() for _ in range(10)]
map = {i: names[i] for i in range(len(names))}
graph = nx.relabel_nodes(G=graph, mapping=map, )
nx.draw_networkx(G=graph)
```

```python
nx.clustering(G=graph)
nx.betweenness.betweenness_centrality(G=graph)
```

### Eigenvector centrality is how well connected a node is based on the amount of its connections' connections
- it's synonymous if not exactly what Zimbabweans mean by _'links'_
- in the above graph, the top shatta is <ins>Christoper Parsons</ins>

```python
nx.eigenvector_centrality(G=graph)
```
